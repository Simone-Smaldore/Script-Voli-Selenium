### Scraping dei dati 

```python
class Volo:
    
  def __init__(self, arrival_city, departure_city, start_date, end_date, price):
    self.id = f'{departure_city}-{arrival_city}-{start_date}-{end_date}-{price}'
    self.arrival_city = arrival_city
    self.departure_city = departure_city
    self.start_date = start_date
    self.end_date = end_date
    self.price = price
    
  def __eq__(self, other): 
    if not isinstance(other, Volo):
        return NotImplemented
    return self.id == other.id 

  def __repr__(self):
    return f'Volo da {self.departure_city} per {self.arrival_city}: {self.price}€   {self.start_date} - {self.end_date}'
    
    
def get_volo_by_div(element_div)-> Volo:
    CLASS_HIDDEN_LI = 'eLNT1d'
    CLASS_CITY = 'W6bZuc'
    CLASS_DATE = 'xyc80b'
    CLASS_PRICE = 'MJg7fb'
    
    if CLASS_HIDDEN_LI in element_div.get_attribute("class"):
        return
    element_arrival_city = element_div.find_element(By.CLASS_NAME, CLASS_CITY).text
    element_date = element_div.find_element(By.CLASS_NAME, CLASS_DATE).text
    element_price = element_div.find_element(By.CLASS_NAME, CLASS_PRICE).text.split(' ')[0]
    (start_date, end_date) = get_date_interval_from_string(element_date)
    return Volo(element_arrival_city, FIRST_CITY, start_date, end_date, element_price)

def get_date_interval_from_string(date: str):
    date_split = date.split('–')
    if date_split[0].isnumeric():
        month = date_split[1].split(' ')[1]
        month = MONTH_MAP[month]
        start_date = datetime.datetime(TODAY.year, month, int(date_split[0]))
        end_date = datetime.datetime(TODAY.year, month, int(date_split[1].split(' ')[0]))
    else:
        start_month = date_split[0].split(' ')[1]
        start_month = MONTH_MAP[start_month]
        end_month = date_split[1].split(' ')[1]
        end_month = MONTH_MAP[end_month]
        start_date = datetime.datetime(TODAY.year, start_month, int(date_split[0].split(' ')[0]))
        end_date = datetime.datetime(TODAY.year, end_month, int(date_split[1].split(' ')[0]))

    if start_date < TODAY:
        start_date = start_date.replace(year=TODAY.year + 1)
    if end_date < TODAY:
        end_date = end_date.replace(year=TODAY.year + 1)
        
    start_date = start_date.strftime("%d/%m/%y")
    end_date = end_date.strftime("%d/%m/%y")
    return start_date, end_date
    
```

```python
def add_voli_to_list(voli):
    try:
        X_PATH_LIST ='/html/body/c-wiz[2]/div/div[2]/div/c-wiz/div[2]/div/div/div[1]/main/div/div[2]/div/ol'
        destination_list = driver.find_element(By.XPATH, X_PATH_LIST)
        destination_div =  destination_list.find_elements(By.TAG_NAME, 'li')
    except:
        return

    cont_eccezione = 0
    for element_div in destination_div:
        try:
            volo = get_volo_by_div(element_div)
            if volo is None:
                cont_eccezione += 1
                continue
            if volo not in voli:
                voli.append(volo)
                print(f'Volo per {volo.arrival_city} al prezzo di {volo.price} euro in data {volo.start_date}/{volo.end_date}')
            cont_eccezione = 0
        except Exception as e:
            print(e)
            cont_eccezione += 1
            if cont_eccezione > 10:
                break    
```
